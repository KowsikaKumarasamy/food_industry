```python
#### visualization
```

```python
diet_pie = df['diet']. value_counts(). reset_index()
plt.figure(figsize = (5,5))

pie = plt.pie(diet_pie['count'], labels = diet_pie['diet'],
             autopct = '%1.1f%%',
             startangle = 90,
             pctdistance = 0.5)
plt.title('Vegetarian vs Non_Vegetarian recipes')
```

```python
region = df['region'].value_counts()
plt.figure(figsize = (10,5))
sns.barplot (x = region.index, y = region.values, palette = 'Greens_r')
plt.title('Region wise distribution of dishes')
plt.xlabel('Region')
plt.ylabel('No.of Dishes')
```

```python
## by flavor
flavor = df['flavor_profile'].value_counts().reset_index()
plt.figure(figsize=(10,6))

plots = sns.barplot(x = flavor['flavor_profile'], y = flavor['count'], palette = 'Greens_r')
for bar in plots.patches:
    plots.annotate(int(bar.get_height()),
                  (bar.get_x() + bar.get_width()/2, bar.get_height()),
                  ha = 'center', va = 'center',
                  size = 12, xytext=(0,8),
                  textcoords = 'offset points')

plt.title('Flavor distribution of dishes')
plt.xlabel('Flavor')
plt.ylabel('No.of dishes')
```

```python
## prep_time and cook_time
df_spicy = df[df['flavor_profile'] == 'spicy']
time = ['cook_time', 'prep_time']
df_plot = df_spicy.sort_values(time, ascending = False).head(10)
plots = df_plot.plot.barh(x = 'name', y = time, color = ['orange','green'],
                         title = 'Prep time and Cook Time comparision for Top 10 spicy dishes', xlabel = 'Time(in minutes)', ylabel = 'Dish name',
                         figsize = (8,8))

for bar in plots.patches :
    plots.annotate(int(bar.get_width()),
                      (bar.get_width(), bar.get_y() + bar.get_height()/8), fontsize = 10)

```

```python
df_sweet = df[df['flavor_profile'] == 'sweet']
time = ['cook_time', 'prep_time']
df_plot = df_sweet.sort_values(time, ascending = False).head(10)
plots = df_plot.plot.barh(x = 'name', y = time, color = ['orange','green'],
                         title = 'Prep time and Cook Time comparision for Top 10 Sweet dishes', xlabel = 'Time(in minutes)', ylabel = 'Dish name',
                         figsize = (8,8))

for bar in plots.patches :
    plots.annotate(int(bar.get_width()),
                      (bar.get_width(), bar.get_y() + bar.get_height()/8), fontsize = 10)
```

```python
flavors = ['sweet','spicy','sour','bitter']
time = ['cook_time', 'prep_time']

fig,axes = plt.subplots(2, 2, figsize=(8,8))
axes = axes.flatten()

for i,flavor in enumerate(flavors):
    df_flavor = df[df['flavor_profile'] == flavor]
    df_plot = df_flavor.sort_values(time, ascending=False).head(10)
    ax = axes[i]
    df_plot.plot.barh(x='name', y=time, color=['orange', 'green'], ax=ax,
                      title=f'{flavor.capitalize()} Dishes Comparison',
                      xlabel='Time (in minutes)', ylabel='Dish name', figsize=(8, 8))

    for bar in ax.patches:
        ax.annotate(int(bar.get_width()), 
                    (bar.get_width(), bar.get_y() + bar.get_height()/8), fontsize=10)

plt.tight_layout()
plt.show()
```

```python
time_plots = ['prep_time','cook_time','total_time']
plt.figure(figsize=(20, 30))

for i, feature in enumerate(time_plots, start=1):
    plt.subplot(4, 2, i)
    plt.tight_layout(pad=5.0) 
    df_plot = df_spicy.sort_values(feature, ascending = False).head(10)
    plots = sns.barplot(x=df_plot[feature],y=df_plot['name'], palette='Greens_r')
    for bar in plots.patches:
        plots.annotate(int(bar.get_width()), 
                        (bar.get_width(), bar.get_y() + bar.get_height() / 2))
    plt.title("Top 10 spicy dishes with highest " + feature)
    plt.xlabel(feature + "(minutes)")
    plt.ylabel("Dish Name")
```
